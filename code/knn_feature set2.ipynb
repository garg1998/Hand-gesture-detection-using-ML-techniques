```python
#loading of dataset
import pandas as pd
features=pd.read_csv("window_feat.csv")
features.columns
```

```python
#dividing dataset into input and target i.e. x and y
feat=features[['IEMG_1', 'IEMG_2', 'MAV_1', 'MAV_2', 'SSI_1', 'SSI_2', 'VAR_1',
       'VAR_2', 'RMS_1', 'RMS_2', 'WL_1', 'WL_2', 'AAC_1', 'AAC_2', 'ZC_1',
       'ZC_2', 'WA_1', 'WA_2', 'SSC_1', 'SSC_2', 'LD_1', 'LD_2', 'DSDAV_1',
       'DASDV_2', 'EMAV_1', 'EMAV_2', 'EWL_1', 'EWL_2', 'MAV1_1', 'MAV1_2',
       'MAV2_1', 'MAV2_2', 'MFL_1', 'MFL_2', 'MYOP_1', 'MYOP_2', 'FR_2',
       'FR_1', 'TTP_1', 'TTP_2', 'MNP_1', 'MNP_2', 'MFN_1', 'MFN_2', 'MFD_1',
       'MFD_2', 'MMFD_2', 'MFMD_1', 'MMFN_1', 'MMFN_2']]
import numpy as np
x=feat
y=np.asarray(features['movement'])
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
scalar= MinMaxScaler()
x=scalar.fit_transform(x)
#PCA implementation
from sklearn.decomposition import PCA
pca=PCA(n_components=None,svd_solver='auto')
PrincipalComponents=pca.fit_transform(x)
x=PrincipalComponents
```

```python
#splitting of data into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
#training of knn classifier
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(x_train, y_train)
pred = classifier.predict(x_test)
```

```python
#classification report
from sklearn.metrics import classification_report 
print(classification_report(y_test,pred))
```

```python
#shuffle split cross validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=96)
cross_val_score(classifier, x, y, cv=cv).mean()
```
